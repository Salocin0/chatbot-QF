```python
from transformers import BertTokenizer, BertForQuestionAnswering
import torch
questions = [
'¿cual es la ciudad de las luces?',
'¿cuales son los monumentos emblematicos de paris?',
'¿cuantos distritos tiene paris?',
'¿porque es famosa francia?',
'¿cual es el idioma oficial?',
'¿cual es la moneda oficial?',
'¿cual es el clima?',
'¿cual es la poblacion de argentina?'
]

context = 'París, la capital de Francia, es una ciudad icónica conocida por su belleza arquitectónica y rica historia. Situada a orillas del río Sena, esta metrópolis alberga monumentos emblemáticos como la Torre Eiffel, el Louvre y la Catedral de Notre-Dame. Con una población de alrededor de 2.15 millones de habitantes, París es un importante centro cultural y turístico. La ciudad es famosa por su gastronomía, particularmente por sus exquisitos croissants, baguettes y la renombrada alta cocina francesa. El idioma oficial es el francés y la moneda es el euro. El clima parisino varía a lo largo del año, con veranos templados e inviernos frescos. París también es reconocida por su moda y es sede de numerosas casas de moda de renombre internacional. La ciudad se compone de veinte distritos, cada uno con su propio carácter distintivo y encanto. París, además de ser una potencia cultural, es un importante centro económico, con un enfoque en sectores como la moda, el turismo, las finanzas y la tecnología.'
model_directory = "./models/1.2.25/" #modelo final
model = BertForQuestionAnswering.from_pretrained(model_directory)
tokenizer = BertTokenizer.from_pretrained('MMG/bert-base-spanish-wwm-cased-finetuned-squad2-es')

max_length = 200

for question in questions:
    answers = []
    # Se tokeniza la pregunta
    inputs = tokenizer(question, context, return_tensors="pt", max_length=max_length, truncation=True)
    start_scores, end_scores = model(**inputs).values()

    # Se obtienen los índices de inicio y fin de la respuesta
    start_index = torch.argmax(start_scores)
    end_index = torch.argmax(end_scores) + 1

    # Se obitenen las respuestas a partir del contexto
    answer = tokenizer.decode(inputs["input_ids"][0][start_index:end_index])

    # Se mejora la respuesta eliminando los [CLS] y [SEP]
    cleaned_answer = answer.replace('[CLS]', '').replace('[SEP]', '').strip()
    answers.append(cleaned_answer)
        

    final_answer = " ".join(answers)

    if not final_answer.strip():  # Si la respuesta está vacía
        print(f"Pregunta: {question}\nRespuesta: No puedo proporcionarte una respuesta, por favor dirige tu consulta a quickfoodcontacto@gmail.com")
    else:  # Si la respuesta no está vacía
        print(f"Pregunta: {question}\nRespuesta: {final_answer}")
    print("\n")
```
