```python
from transformers import BertTokenizer, BertForQuestionAnswering
import torch
questions = [
"¿que es quickfood?"
,"¿Quickfood permite la actualización en tiempo real del menú para los Encargados de Puestos de Comida?"
,"¿Los Productores de Eventos pueden gestionar varios puestos de comida simultáneamente en Quickfood?"
,"¿con que celulares es compatible quickfood?"
,"¿funciona en ios?"
,"¿Quickfood coordina pedidos para eventos masivos y privados?"
,"¿Los Productores de Eventos pueden monitorear el estado y congestión de los pedidos en Quickfood?"
,"¿Los Encargados de Puestos de Comida pueden gestionar pagos y transacciones a través de Quickfood?"
,"¿Quickfood permite la personalización de menús para eventos específicos?"
,"¿Los Productores de Eventos pueden comunicarse directamente con los Encargados de Puestos de Comida en Quickfood?"
,"¿Los Encargados de Puestos de Comida reciben notificaciones en tiempo real sobre los pedidos en Quickfood?"
,"¿Cómo puede un Consumidor registrarse como Repartidor en Quickfood?"
]

context = 'Quickfood, una aplicación versátil y eficiente, compatible con dispositivos Android, pero no es compatible con IOS. brinda soluciones a diversos usuarios en el ámbito de eventos y gastronomía. ¿Te preguntas si los Encargados de Puestos de Comida pueden actualizar su menú en tiempo real? ¡Claro! Quickfood les ofrece esta capacidad, permitiéndoles personalizar y variar su oferta para satisfacer las preferencias de los Consumidores. Si eres Productor de Eventos, te preguntarás si puedes gestionar múltiples puestos de comida a través de Quickfood. ¡La respuesta es sí! La aplicación te brinda la capacidad de coordinar y administrar eficientemente varios puestos para eventos masivos o privados. Además puedes comunicarte directamente con los Encargados de Puestos de Comida. Los Repartidores pueden acceder a la ubicación exacta del consumidor para realizar entregas directas a sus asientos o lugares designados.Para los Consumidores, Quickfood ofrece la comodidad de realizar pedidos desde cualquier asiento en el evento, brindando una experiencia eficiente y cómoda. Además, se permite realizar un seguimiento en tiempo real de los pedidos, desde la realización hasta la entrega. Además, si eres Consumidor puedes realizar precompras sin necesidad de tener entradas. Los Encargados de Puestos de Comida reciben notificaciones en tiempo real sobre los pedidos y pueden gestionar pagos y transacciones fácilmente a través de Quickfood, facilitando una experiencia integrada y conveniente. Además, los Productores de Eventos pueden visualizar el estado y congestión de los pedidos en tiempo real. En cuanto a los métodos de pago, Quickfood acepta Mercado Pago para transacciones seguras y convenientes. Para Repartidores, la aplicación proporciona información detallada sobre los pedidos pendientes y las ubicaciones de entrega. Un consumidor puede ser repartidor y presentarse en un evento a realizar pedidos. Si te interesa ser Repartidor en Quickfood debes registrarte primero como Consumidor y proporcionarnos tus datos. Para ser Productor o Encargados solo te solicitaremos datos relacionados a tu empresa o negocio.Quickfood se adapta a diversas situaciones, permitiendo a los usuarios personalizar menús para eventos específicos, ya sea que sean Productores de Eventos, Encargados de Puestos de Comida o Consumidores. La aplicación también se preocupa por la situación de eventos imprevistos, ofreciendo reembolsos en caso de suspensión o cancelación. Además, proporciona créditos equivalentes a precompras para eventos suspendidos, demostrando su compromiso con los usuarios. Para garantizar la seguridad de la información del usuario, Quickfood emplea medidas de encriptación avanzadas y protocolos de seguridad. Además, cuenta con un equipo de soporte técnico para resolver cualquier inconveniente o inquietud de los usuarios.'
model_directory = "./models/1.2.25/"
model = BertForQuestionAnswering.from_pretrained(model_directory)
#model = BertForQuestionAnswering.from_pretrained('MMG/bert-base-spanish-wwm-cased-finetuned-squad2-es')
tokenizer = BertTokenizer.from_pretrained('MMG/bert-base-spanish-wwm-cased-finetuned-squad2-es')

max_length = 200

for question in questions:
    answers = []
    # Se tokeniza la pregunta
    inputs = tokenizer(question, context, return_tensors="pt", max_length=max_length, truncation=True)
    start_scores, end_scores = model(**inputs).values()

    # Se obtienen los índices de inicio y fin de la respuesta
    start_index = torch.argmax(start_scores)
    end_index = torch.argmax(end_scores) + 1

    # Se obitenen las respuestas a partir del contexto
    answer = tokenizer.decode(inputs["input_ids"][0][start_index:end_index])

    # Se mejora la respuesta eliminando los [CLS] y [SEP]
    cleaned_answer = answer.replace('[CLS]', '').replace('[SEP]', '').strip()
    answers.append(cleaned_answer)
        

    final_answer = " ".join(answers)

    if not final_answer.strip():  # Si la respuesta está vacía
        print(f"Pregunta: {question}\nRespuesta: No puedo proporcionarte una respuesta, por favor dirige tu consulta a quickfoodcontacto@gmail.com")
    else:  # Si la respuesta no está vacía
        print(f"Pregunta: {question}\nRespuesta: {final_answer}")
    print("\n")
```
